# for SB data :  check with meter 

```python
import pandas as pd

first_value = None

# Initialize an empty column
warehouse_Data['matched_udc_id'] = None

for value in warehouse_Data['Meter No.']:
    first_value = value
    
    # Find the index in warehouse_Data
    index_of_match_warehouse = warehouse_Data.index[warehouse_Data['Meter No.'] == first_value].tolist()[0]
    print(f"index_of_match : {index_of_match_warehouse}")

    if first_value in assest_initiation_invent_sb['MFG_SERIAL_NUM'].tolist():
      
        print("Meter match")
        print(f"Meter No.: {value}")
        
        # Get the index of the matching value
        index_of_match_assest_initi_invent_sb = assest_initiation_invent_sb.index[assest_initiation_invent_sb['MFG_SERIAL_NUM'] == value].tolist()[0]
        print(f"index_of_match_assest_initi_invent_sb : {index_of_match_assest_initi_invent_sb}")

        # Print the corresponding 'UDC_ID'
        METER_matched_udc_id = assest_initiation_invent_sb.loc[index_of_match_assest_initi_invent_sb, 'UDC_ID']
        METER_matched_IP =  assest_initiation_invent_sb.loc[index_of_match_assest_initi_invent_sb, 'NETWORK_ID']
        METER_matched_meter_status =  assest_initiation_invent_sb.loc[index_of_match_assest_initi_invent_sb, 'Status']
        METER_matched_utility =  assest_initiation_invent_sb.loc[index_of_match_assest_initi_invent_sb, "DECODE(D.ORG_ID,52,'SBPDCL',102,'NBPDCL')"]
        print(f"index_of_match_assest_initi_invent_sb: {index_of_match_assest_initi_invent_sb} ,  Matching UDC_ID: {METER_matched_udc_id}")
        print(f"Matching UDC_ID: {METER_matched_IP}")
        print(f"Matching UDC_ID: {METER_matched_meter_status}")
        print(f"Matching UDC_ID: {METER_matched_utility}")
        
        # Update the 'matched_udc_id' column in warehouse_Data
        warehouse_Data.at[index_of_match_warehouse, 'METER_matched_udc_id'] = METER_matched_udc_id
        warehouse_Data.at[index_of_match_warehouse, 'METER_matched_IP'] = METER_matched_IP
        warehouse_Data.at[index_of_match_warehouse, 'METER_matched_meter_status'] = METER_matched_meter_status
        warehouse_Data.at[index_of_match_warehouse, 'METER_matched_utility'] = METER_matched_utility
        print("*" * 70)

# Optionally, you can save the updated DataFrame to a new CSV file
# warehouse_Data.to_csv('output_data.csv', index=False)
```
