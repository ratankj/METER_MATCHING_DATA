# FINAL : - SIM SET CHECK STATUS 

```python


def fun_sim_status_check(row):
    if (row['IMSI'] == row['METER_matched_udc_id']) and (row['final hes mapping'] == row['METER_matched_utility']):
        return "SAME SET IN SAME UTILITY"
    elif (row['IMSI'] != row['METER_matched_udc_id']) and (row['final hes mapping'] == row['METER_matched_utility']):
        return "DIFFERENT SET IN SAME UTILITY"
    elif (row['IMSI'] == row['METER_matched_udc_id']) and (row['final hes mapping'] != row['METER_matched_utility']):
        return "SAME SET IN DIFFERENT UTILITY"
    elif (row['IMSI'] != row['METER_matched_udc_id']) and (row['final hes mapping'] != row['METER_matched_utility']):
        return "DIFFERENT SET IN DIFFERENT UTILITY"

# Apply the function to create a new column
warehouse_Data['SIM SET CHECK STATUS'] = warehouse_Data.apply(fun_sim_status_check, axis=1)

```

# TRUE - FALSE COLUMN

```python
def fun_check_true_false__METER_SIM(row):
    if row['IMSI']==row['METER_matched_udc_id']:
        return True
    else:
        return False
    
warehouse_Data["METER_SIM"] =  warehouse_Data.apply(fun_check_true_false__METER_SIM, axis=1)
```

```python
def fun_check_true_false_METER_ORG(row):
    if row['METER_matched_utility']==row['final hes mapping']:
        return True
    else:
        return False
    
warehouse_Data["METER_ORG"] =  warehouse_Data.apply(fun_check_true_false_METER_ORG, axis=1)
```

```python
def fun_check_true_false_SIM_ORG(row):
    if row['IMSI_matched_utility']==row['final hes mapping']:
        return True
    else:
        return False
    
warehouse_Data["SIM_ORG"] =  warehouse_Data.apply(fun_check_true_false_SIM_ORG, axis=1)
```

```python
def fun_check_true_false_METER_IP(row):
    if row['METER_matched_IP']==row['IP']:
        return True
    else:
        return False
    
warehouse_Data["METER_IP"] =  warehouse_Data.apply(fun_check_true_false_METER_IP, axis=1)
```

# SIM SET CHECK STATUS RECHECK COLUMN

```python
def check_set_and_utility(row):
    if row['METER_SIM'] == True and (row['METER_ORG'] and row["SIM_ORG"] == True) :
        return "SAME SET IN SAME UTILITY"
    elif row['METER_SIM'] == True and (row['METER_ORG'] and row["SIM_ORG"] == False) :
        return "SAME SET IN DIFFERENT UTILITY"
    elif row['METER_SIM'] == False and (row['METER_ORG'] and row["SIM_ORG"] == True) :
        return "DIFFERENT SET IN SAME UTILITY"
    elif row['METER_SIM'] == False and (row['METER_ORG'] and row["SIM_ORG"] == False) :
        return "DIFFERENT SET IN DIFFERENT UTILITY"


warehouse_Data["SIM SET CHECK STATUS RECHECK"] =  warehouse_Data.apply(check_set_and_utility, axis=1)
```

# ACTION TAKEN COLUMN


```python

def fun_action_taken_case_I(row):
    if (row['METER_matched_meter_status'] == "Installed")  & (row['IMSI_matched_meter_status'] == "Installed"):
        return "SET IS IN INSTALLED STAGE"
    elif (row['METER_matched_meter_status'] != "Installed")  & (row['IMSI_matched_meter_status'] == "Installed"):
        return "IMSI IS IN INSTALLED STAGE"
    elif (row['METER_matched_meter_status'] == "Installed")  & (row['IMSI_matched_meter_status'] != "Installed"):
        return "METER IS IN INSTALLED STAGE"
    
warehouse_Data["ACTION TAKEN"] =  warehouse_Data.apply(fun_action_taken_case_I, axis=1)
```

```python
warehouse_Data.head(1)
```

```python
def fun_action_taken_case_II(row):

    # set 1 and row['METER_IP'] == True
    if (row['METER_SIM'] == True) and (row['METER_ORG'] and row["SIM_ORG"] == True) and (row['METER_IP'] == True):
        return "METER , SIM LINK FILE REQUIRED"
    
    # set 1 and row['METER_IP'] == False 
    elif (row['METER_SIM'] == True) and (row['METER_ORG'] and row["SIM_ORG"] == True) and (row['METER_IP'] == False) :
        return "SIM AND METER LINK FILE REQUIRED"
    
    
        
warehouse_Data["ACTION TAKEN"] =  warehouse_Data.apply(fun_action_taken_case_II, axis=1)
```

```python
def fun_action_taken_case_III_and_case_IV(row):
    
    # set 2  "SAME SET IN DIFFERENT UTILITY"
    if (row['METER_SIM'] == True) and (row['METER_ORG'] and row["SIM_ORG"] == False) :
        return "METER,SIM,METER-SIM LINK FILE REQUIRED"
    
    #set 3  "DIFFERENT SET IN SAME UTILITY"
    
    elif (row['METER_SIM'] == False) and (row['METER_ORG'] and row["SIM_ORG"] == True) :
        return "SIM AND METER SIM LINK FILE REQUIRED"
    
    # set 4  "DIFFERENT SET IN DIFFERENT UTILITY"
    elif (row['METER_SIM'] == False) and (row['METER_ORG'] and row["SIM_ORG"] == False) :
        return "METER,SIM,METER-SIM LINK FILE REQUIRED"


warehouse_Data["ACTION TAKEN"] =  warehouse_Data.apply(fun_action_taken_case_III_and_case_IV, axis=1)
```
