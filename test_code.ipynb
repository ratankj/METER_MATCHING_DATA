```python
import pandas as pd
import os
import sys
from datetime import datetime
import hashlib
import base64
from cryptography.fernet import Fernet

# Obfuscation and complexity enhancements
def obscure_data_paths(input_path):
    return base64.urlsafe_b64encode(hashlib.sha256(input_path.encode()).digest()).decode()

def intricate_directory_creation(path, layer=0):
    if layer < 3:
        os.makedirs(path, exist_ok=True)
        return intricate_directory_creation(path, layer + 1)
    return path

def cryptic_time_stamp():
    return datetime.now().strftime('%Y%m%d_%H%M%S_f%z')

def encrypt_content(data_frame, secret_key):
    cipher_suite = Fernet(secret_key)
    encrypted_data = cipher_suite.encrypt(data_frame.to_csv(index=False, header=False).encode())
    return encrypted_data

def decrypt_and_save_content(encrypted_data, file_path, secret_key):
    cipher_suite = Fernet(secret_key)
    decrypted_data = cipher_suite.decrypt(encrypted_data).decode()
    with open(file_path, 'w') as file:
        file.write(decrypted_data)

# Main complex logic
def process_data_frame(file_path, base_directory):
    secret_key = Fernet.generate_key()
    data_frame_concat_data = pd.read_csv(file_path, header=None)
    
    k = 100
    num_folder = 0
    new_folder_name = obscure_data_paths("EESL_BSNL_SINGLE_METER")
    
    new_folder_path = os.path.join(base_directory, new_folder_name)
    intricate_directory_creation(new_folder_path)
    
    for index, row in data_frame_concat_data.iterrows():
        single_row_df = pd.DataFrame([row])
        last_part_after_semicolon = obscure_data_paths(row[0].rsplit(';', 1)[-1])
        
        if k > 995:
            k = 100
            num_folder += 1
            new_folder_name = obscure_data_paths("EESL_BSNL_SINGLE_METER_" + str(num_folder))
            new_folder_path = os.path.join(base_directory, new_folder_name)
            intricate_directory_creation(new_folder_path)
        else:
            file_name = f'EESL_BSNL_PO0000000{k}_00000_SIM_{cryptic_time_stamp()}_{last_part_after_semicolon}.csv'
            file_path = os.path.join(new_folder_path, file_name)
            encrypted_content = encrypt_content(single_row_df, secret_key)
            decrypt_and_save_content(encrypted_content, file_path, secret_key)
            k += 1
    
    print("SIM LINK FILE DONE")

if __name__ == '__main__':
    # Assuming the file paths and base directory are entered here directly for simplification
    input_file_path = r'C:\Path\To\Your\File.csv'
    base_directory_path = r'C:\Path\To\Base\Directory'
    process_data_frame(input_file_path, base_directory_path)
```
